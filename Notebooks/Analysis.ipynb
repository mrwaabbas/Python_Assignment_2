```python
import pandas as pd
```

```python
df = pd.read_csv("../data/NYC Evictions_rawdata.csv")
df
```

```python
clean_df = df[["BOROUGH", "Eviction Postcode", "Executed Date"]]
```

```python
year = clean_df["Executed Date"].dt.year
clean_df = pd.concat([clean_df, year.rename("year")], axis=1)
```

```python
print(clean_df.head())
```

```python
clean_df = clean_df.drop(columns=["Executed Date"])

print(clean_df.head())
```
